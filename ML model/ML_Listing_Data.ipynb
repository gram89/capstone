```python
import pandas as pd
import numpy as np
import pandas as pd
from pathlib import Path
from collections import Counter
```

```python
df = pd.read_csv("Listing_cleaned.csv")
df.head()
```

```python
#separate the other attributes from the predicting attribute
X = df[['Beds', 'Baths', 'City']]

#separate the predicting attribute into Y for model training 
y = df['Price']
```

```python
X.describe()
```

```python
# Check the balance of our target values
y.value_counts()
```

```python
# handle categorical variable
X = pd.get_dummies(X,drop_first=True)
X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

print("----------------------------------------")
print("Balance of the independent variables:\n")
print(f"X_train: {Counter(X_train.shape)}")
print(f"X_test {Counter(X_test.shape)}")

print("----------------------------------------")
print("Balance of the target variables:\n")
print(f"y_train: {Counter(y_train.shape)}")
print(f"y_test {Counter(y_test.shape)}")
print("----------------------------------------")
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# Fit the model to the training data, and calculate the scores for the training and testing data.

### BEGIN SOLUTION
model.fit(X_train, y_train)
training_score = model.score(X_train, y_train)
testing_score = model.score(X_test, y_test)

### END SOLUTION 

print(f"Training Score: {training_score}")
print(f"Testing Score: {testing_score}")
```
