```python
#import dependencies

import pandas as pd
from time import time, sleep
from requests import Session
from bs4 import BeautifulSoup as soup
import time
```

```python
#initiating the session
main_session = Session()
table = []
url = "https://barrie.listing.ca/real-estate-price-history.htm"
resp = main_session.get(url)

historical_page = soup(resp.content, 'html.parser')

print("Classes of each table:")
for table in historical_page.find_all("table"):
    print(table.get("class"))
    print(table.get("style"))
    
```

```python
#finding all the tables with the relevant information
tables = historical_page.find_all("table",style = "width:100%;", cellspacing = "5")
row = tables[3].find_all("tr")#.find_all("td")
#row[2].find_all("td")
row
```

```python
df = pd.DataFrame(columns = ["Month","Sold","List_price_avg","Sold_price_avg","above_below","monthly_change","monthly_change_per","days_on_mar"])

for i in range(0,len(tables)):
    row = tables[i].find_all("tr")
    for j in range(0,len(row)):
        columns = row[j].find_all("td")
        
        if (columns !=[]):
            a = columns[0].text.strip()
            b = columns[1].text.strip()
            c = columns[2].text.strip()
            d = columns[3].text.strip()
            e = columns[4].text.strip()
            f = columns[5].text.strip()
            g = columns[6].text.strip()
            h = columns[7].text.strip()
            
            df = df.append({"Month": a, "Sold": b, "List_price_avg": c, "Sold_price_avg": d, "above_below": e, "monthly_change": f, "monthly_change_per": g, "days_on_mar": h}, ignore_index=True)
```
