```python
#import dependencies

import pandas as pd
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter

# initialize Nominatim API 
geolocator = Nominatim(user_agent="geoapiExercises")
geocode = RateLimiter(lambda query: geolocator.geocode("%s, Ontario" % query),min_delay_seconds=1)
```

```python
#get the source file

df2 = pd.read_csv("capstone/listing_cleaned.csv")
df2.head()
```

```python
address = df2["Address"]
address[1090]
postal_c=[]
```

```python

for i in range (0, len(address)):
    place = address[i]
    location = geolocator.geocode(geocode(place))
    print(location)
    if location is None:
        print(f"no location for {address[i]}")
        postal_code = 000000
    else: 
        postal= str(location).split(", ")
        postal_code = postal[-2]
    postal_c.append ({"Postal_Code": postal_code})
```
